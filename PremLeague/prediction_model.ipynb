```python
# clean the data
matches.dtypes 
```

```python
matches["date"] = pd.to_datetime(matches["date"])
matches.dtypes 
```

```python
##converting all objects to int or float to be processed by the machine learning software
matches["date"] = pd.to_datetime(matches["date"])
matches["h/a"] = matches["venue"].astype("category").cat.codes ## converting venue to a home (1) or away (0) number
matches["opp"] = matches["opponent"].astype("category").cat.codes ## converting opponents to a number
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int") ## converting hours to number in case a team plays better at a certain time
matches["day"] = matches["date"].dt.dayofweek ## converting day of week of game to a number
```

```python
## setting a win to the value 1
matches["target"] = (matches["result"] == "W").astype("int") 
matches
```

```python
from sklearn.ensemble import RandomForestClassifier

# training 100 decision trees on the data with 10 samples in each tree
rf = RandomForestClassifier(n_estimators=100, min_samples_split=10, random_state=1)
```

```python
train = matches[matches["date"] < "2022-01-01"]
test = matches[matches["date"] > "2022-01-01"]
```

Here we're training the model so that when we test it on data in the future, it has a better chance of predicting the outcome of the game because of the model having been trained.


```python
predictors = ["opp", "h/a", "day", "hour"]
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(test["target"], preds) ## testing accuracy
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual=test["target"], prediction=preds))
pd.crosstab(index=combined["actual"], columns=combined["prediction"])
```

Here we see that when we predicted a loss/draw, we were correct 142 times and wrong 79 times which is good for this accuracy_score based off 100 decision trees.

But when we predicted a win, we were correct 30 times and wrong 25 times which is almost 50% which is too close for us to be confident in the model.

Thus we'll use precision_score increase the accuracy of the model.

```python
from sklearn.metrics import precision_score
precision_score(test["target"], preds)
```

We'll create more predictors to improve the accuracy of the model.

Let's split our matches dataframe by teams because we want to compute rolling averages for each team which will allow us to see how well a team has been doing in their past games based on shot attempts, goals, etc.


```python
grouped_matches = matches.groupby("team")
group = grouped_matches.get_group("Chelsea")
group.head()
```

We need to compute the rolling averages for Liverpool and basically say that if we're in matchweek 4, how did Liverpool do in the past 3 weeks.


```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed="left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling = matches_rolling.droplevel('team') ## dropping extra index level
matches_rolling.index = range(matches_rolling.shape[0]) ## adding new index
matches_rolling
```

Now that we have new predictors, we can train the model again.


```python
def make_predictions(data, predictors, model):
    train = data[data["date"] < "2022-01-01"]
    test = data[data["date"] > "2022-01-01"]
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual = test["target"], prediction=preds))
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols, rf)
precision
```

```python
combined
```

Our precision scored has increased from 45% to 57% which is a good improvement.

But there are still come mispredictions as we can see from the combined but it doesn't tell us which team actually played in each match meaning we're unable to see if we're mispredicting and team.

To fix that we'll add that information to the combined dataframe with combined.merge


```python
combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index = True, right_index = True)
combined
```

We noticed that the team name and the opponent name are not listed as the same. For example, team = Wolverhampton Wanderers and opponent = Wolves.

To make this consistent, we can use a dictionary and then the pandas map function with that dictionary.


```python
class MissingDict(dict): ## creating a class that inherits from the dictionary class
    __missing__ = lambda self, key: key ## in case a team name is missing

map_values = {
    "Brighton and Hove Albion": "Brighton",
    "Manchester United": "Manchester Utd",
    "Tottenham Hotspur": "Tottenham", 
    "West Ham United": "West Ham", 
    "Wolverhampton Wanderers": "Wolves"
}

mapping = MissingDict(**map_values)
mapping["West Ham United"]
```

We can use these new and fixed names in the pandas map method.

```python
combined["new_team"] = combined["team"].map(mapping)
combined
```

Look at the dataframe and see that there is a team and the opponent they play. then the opponent versus the team. 
So basically x versus y and y versus x are in the dataframe entailing why it's best to merge it with itself.


```python
merged = combined.merge(combined, left_on = ["date", "new_team"], right_on = ["date", "team"])
merged
```
